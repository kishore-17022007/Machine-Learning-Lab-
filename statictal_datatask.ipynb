```python
Corr = data.corr(numeric_only=True)
print(Corr)
Corr = pd.DataFrame(Corr)
Corr.to_csv('out.csv')
```

```python
f,ax = plt.subplots(figsize=(9, 9))
sns.heatmap(Corr, annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
f,ax = plt.subplots(figsize=(9, 9))
sns.heatmap(data.cov(numeric_only=True), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```
